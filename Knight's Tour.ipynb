```python
def mergesort(A,left,right):
 # Sort the slice A[left:right]
 if right - left <= 1: # Base case
    return(A[left:right])
 if right - left > 1: # Recursive call
    mid = (left+right)//2
    L = mergesort(A,left,mid)
    R = mergesort(A,mid,right)
    return(merge(L,R))
def merge(A,B):
    C,m,n=([],len(A),len(B))
    (i,j)=(0,0)
    while(i+j<m+n):
        if(i==m):
            C.append(B[j])
            j=j+1
        elif(j==n):
            C.append(A[i])
            i=i+1
        elif(A[i][1]>B[j][1]):
            C.append(A[i])
            i=i+1
        elif(A[i][1]<=B[j][1]):
            C.append(B[j])
            j=j+1
    return C
def neighbour(ax,ay,M,N,marked,flag):
    if(flag==0):
        y=[]
        next_box=[[ax-1,ay-2],[ax-1,ay+2],[ax+1,ay-2],[ax+1,ay+2],[ax+2,ay-1],[ax+2,ay+1],[ax-2,ay-1],[ax-2,ay+1]]
        for i in range(len(next_box)):
            if ((next_box[i][0]>=0 and next_box[i][0]<M) and ((next_box[i][1]>=0 and next_box[i][1]<N)) and is_matrix_marked([[next_box[i][0],next_box[i][1]]],marked)):
                y.append(next_box[i])
        return(len(y))
        
                
    else:
        next_box=[[ax-1,ay-2],[ax-1,ay+2],[ax+1,ay-2],[ax+1,ay+2],[ax+2,ay-1],[ax+2,ay+1],[ax-2,ay-1],[ax-2,ay+1]]
        x=[]
        for i in range(len(next_box)):
            if ((next_box[i][0]>=0 and next_box[i][0]<M) and ((next_box[i][1]>=0 and next_box[i][1]<N))):
                
                x.append([next_box[i],neighbour(next_box[i][0],next_box[i][1],M,N,marked,flag=0)])
        
        neighbour_sorted=mergesort(x,0,len(x))
        return neighbour_sorted
def is_matrix_marked(s,marked):
    return not(marked[s[0][0]][s[0][1]]==1)
def knights_path(M,N,i,j,marked,queue,prop):
    
    marked_matrix(i,j,marked)
    s=[]
    if(not matrixEmpty(marked,prop)):
        s= neighbour(i,j,M,N,marked,flag=1)
        s=list(filter(lambda x: is_matrix_marked(x,marked), s))
        if(is_empty(s)):
            undo_matrix(i,j,marked)
            return False
        if(knights_path(M,N,s[-1][0][0],s[-1][0][1],marked,queue,prop)):
            queue.insert(0,(s[-1][0][0]+1,s[-1][0][1]+1))
            return True
        else:
            undo_matrix(s[-1][0][0],s[-1][0][1],marked)
            s.pop()
            if(is_empty(s)):
                return False
                
            else:
                if(not knights_path(M,N,s[-1][0][0],s[-1][0][1],marked,queue,prop)):
                    undo_matrix(s[-1][0][0],s[-1][0][1],marked)
                    s.pop()
                    return False
                else:
                    queue.insert(0,(s[-1][0][0]+1,s[-1][0][1]+1))
                    return True
                        
                    
    else:
        return True
def marked_matrix(i,j,marked):
    marked[i][j]=1
def undo_matrix(i,j,marked):
    marked[i][j]=0
def matrixEmpty(marked,prop):
    return(marked == prop )
    
def is_empty(s):
    return (len(s)==0)   
    
```

```python
def knightstour(M,N,i,j):
    queue=[]

    if((i>=1 and j>=1) and (i<=M and j<=N)):
        i=i-1
        j=j-1
        prop=[[1 for p in range(N)] for q in range(M)]
        marked=[[0 for p in range(N)] for q in range(M)]
        if(knights_path(M,N,i,j,marked,queue,prop)):
            queue.insert(0,(i+1,j+1))
            return(queue)
            

        else:
            return queue
    else:
        return queue
```
